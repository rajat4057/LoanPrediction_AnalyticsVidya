```python
df['ApplicantIncome']
q1=df['ApplicantIncome'].quantile(q=0.25)
q3=df['ApplicantIncome'].quantile(q=0.75)
iqr=q3-q1
lb=(q1-(1.5*iqr))
ub=(q3+(1.5*iqr))
print("Q1:{},Q3:{},IQR:{},LB:{},UB:{}".format(q1,q3,iqr,lb,ub))
```

```python
list1=['ApplicantIncome','CoapplicantIncome','LoanAmount']
for l in list1:
    q1=df[l].quantile(q=0.25)
    q3=df[l].quantile(q=0.75)
    iqr=q3-q1
    lb=(q1-(1.5*iqr))
    ub=(q3+(1.5*iqr))
    count=0
    for val in df[l]:
        if val>ub:
            count=count+1
    print("Column:{},Q1:{},Q3:{},IQR:{},LB:{},UB:{},Outliers:{}".format(l,q1,q3,iqr,lb,ub,count))
    
```

```python
def loan_term(row):
    if row['Loan_Amount_Term']<=60:
        term="Very_Short_Term"
    elif (row['Loan_Amount_Term']>60) & (row['Loan_Amount_Term']<180):
        term="Short_Term"
    elif (row['Loan_Amount_Term']>=180) & (row['Loan_Amount_Term']<=300):
        term="Long_Term"
    elif row['Loan_Amount_Term']>300:
        term="Very_Long_Term"
    return term
df['Loan_Term']=df.apply(loan_term,axis=1)
```

```python
def ls(row):
    if row['Loan_Status']=="Y":
        loan=1
    else:
        loan=0
    return loan
df_new['LoanStatus']=df_new.apply(ls,axis=1)
```

```python
from sklearn import preprocessing
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
scaled_vals=minmax.fit_transform(df_new[['ApplicantIncome','CoapplicantIncome','LoanAmount']])
```

```python
def feature_scaling(df):
    df -= df.min()
    df /= df.max()
    return df
```

```python
df_new['ApplicantIncome']=feature_scaling(df_new['ApplicantIncome'])
df_new['CoapplicantIncome']=feature_scaling(df_new['CoapplicantIncome'])
df_new['LoanAmount']=feature_scaling(df_new['LoanAmount'])
```
